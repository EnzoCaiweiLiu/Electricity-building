```python
import numpy as np
import pandas as pd
```

```python
#Get data
data_1 = pd.read_csv('data1.csv')
data_2 = pd.read_csv('data2.csv')
```

```python
#Reliability of robots
c_r1 = 0
c_r2 = 0
for i in range(len(data_1)):
    ob1 = data_1['robot1'].iloc[i].split('\'')[1]
    ob2 = data_1['robot2'].iloc[i].split('\'')[1]
    ob_num1 = data_1['robot1'].iloc[i].split(',')[1].split(')')[0]
    ob_num2 = data_1['robot2'].iloc[i].split(',')[1].split(')')[0]
    re_num1 = data_1[ob1].iloc[i]
    re_num2 = data_1[ob2].iloc[i]
    if int(ob_num1) == int(re_num1):
        c_r1 += 1
    if int(ob_num2) == int(re_num2):
        c_r2 +=1
for i in range(len(data_2)):
    ob1 = data_2['robot1'].iloc[i].split('\'')[1]
    ob2 = data_2['robot2'].iloc[i].split('\'')[1]
    ob_num1 = data_2['robot1'].iloc[i].split(',')[1].split(')')[0]
    ob_num2 = data_2['robot2'].iloc[i].split(',')[1].split(')')[0]
    re_num1 = data_2[ob1].iloc[i]
    re_num2 = data_2[ob2].iloc[i]
    if int(ob_num1) == int(re_num1):
        c_r1 += 1
    if int(ob_num2) == int(re_num2):
        c_r2 +=1
print(c_r1/(len(data_1)+len(data_2)))
print(c_r2/(len(data_1)+len(data_2)))

#Robot is 100% reliable
```

```python
#Compute Reliability of Motion Sensors
def compute_motion_sensor(sensor,room):
    acc_t = 0
    acc_f = 0
    all_motion_sen = 0
    all_no_motion_sen = 0
    correct = 0
    for i in range(len(data_1)):
        ob = data_1[sensor].iloc[i]
        ob_room = data_1[room].iloc[i]
        if ob == 'motion':
            all_motion_sen += 1
            if int(ob_room) > 0:
                acc_t += 1
                correct += 1
        else:
            all_no_motion_sen += 1
            if int(ob_room) == 0:
                correct += 1
            if int(ob_room) > 0:
                acc_f += 1
    for i in range(len(data_2)):
        ob = data_2[sensor].iloc[i]
        ob_room = data_2[room].iloc[i]
        if ob == 'motion':
            all_motion_sen += 1
            if int(ob_room) > 0:
                acc_t += 1
                correct += 1
        else:
            all_no_motion_sen += 1
            if int(ob_room) == 0:
                correct += 1
            if int(ob_room) > 0:
                acc_f += 1
    print('TP for',sensor,':',acc_t/all_motion_sen)
    print('FN for',sensor,':',acc_f/all_no_motion_sen)
    print('Accuracy for',sensor,':',correct/(len(data_1)+len(data_2)))
    print('\n')
```

```python
#Motion Censor
compute_motion_sensor('motion_sensor1','r1')
compute_motion_sensor('motion_sensor2','r2')
compute_motion_sensor('motion_sensor3','r3')
compute_motion_sensor('motion_sensor4','r4')
compute_motion_sensor('motion_sensor5','r5')
compute_motion_sensor('motion_sensor6','r6')
compute_motion_sensor('motion_sensor7','r7')
compute_motion_sensor('motion_sensor8','r8')
compute_motion_sensor('motion_sensor9','r9')
compute_motion_sensor('motion_sensor10','r10')
```

```python
#Compute Reliability of Camera Sensors
def compute_camera_sensor(sensor,room):
    correct = 0
    for i in range(len(data_1)):
        ob = data_1[sensor].iloc[i]
        ob_room = data_1[room].iloc[i]
        if ob > 0 and ob_room > 0:
            correct += 1
        elif ob == 0 and ob_room == 0:
            correct += 1

    for i in range(len(data_2)):
        ob = data_2[sensor].iloc[i]
        ob_room = data_2[room].iloc[i]
        if ob > 0 and ob_room > 0:
            correct += 1
        elif ob == 0 and ob_room == 0:
            correct += 1

    print('Accuracy for',sensor,':',correct/(len(data_1)+len(data_2)))
    print('\n')
```

```python
#Camera
compute_camera_sensor('camera1','r1')
compute_camera_sensor('camera2','r4')
compute_camera_sensor('camera3','r8')
compute_camera_sensor('camera4','c3')
```

```python
#Compute Reliability of Door Sensors
def compute_door_sensor(sensor,room1,room2):
    acc_t_r1 = 0
    acc_f_r1 = 0
    acc_t_r2 = 0
    acc_f_r2 = 0
    correct = 0
    all_motion_sen = 0
    all_no_motion_sen = 0
    for i in range(len(data_1)):
        ob = data_1[sensor].iloc[i]
        ob_room1 = data_1[room1].iloc[i]
        ob_room2 = data_1[room2].iloc[i]
        if int(ob) > 0:
            all_motion_sen += 1
            if int(ob_room1) > 0:
                acc_t_r1 += 1
            if int(ob_room2) > 0:
                acc_t_r2 += 1
            if int(ob_room1)> 0 or int(ob_room2)> 0:
                correct += 1
        else:
            all_no_motion_sen += 1
            if int(ob_room1) > 0:
                acc_f_r1 += 1
            if int(ob_room2) > 0:
                acc_f_r2 += 1
            if int(ob_room1) ==  0 or int(ob_room2) == 0:
                correct += 1
    
    for i in range(len(data_2)):
        ob = data_2[sensor].iloc[i]
        ob_room1 = data_2[room1].iloc[i]
        ob_room2 = data_2[room2].iloc[i]
        if int(ob) > 0:
            all_motion_sen += 1
            if int(ob_room1) > 0:
                acc_t_r1 += 1
            if int(ob_room2) > 0:
                acc_t_r2 += 1
            if int(ob_room1)> 0 or int(ob_room2)> 0:
                correct += 1
        else:
            all_no_motion_sen += 1
            if int(ob_room1) > 0:
                acc_f_r1 += 1
            if int(ob_room2) > 0:
                acc_f_r2 += 1
            if int(ob_room1) ==  0 or int(ob_room2) == 0:
                correct += 1
    print('TP for',sensor,'of',room1,':',acc_t_r1/all_motion_sen)
    print('FN for',sensor,'of',room1,':',acc_f_r1/all_no_motion_sen)
    print('TP for',sensor,'of',room2,':',acc_t_r2/all_motion_sen)
    print('FN for',sensor,'of',room2,':',acc_f_r2/all_no_motion_sen)
    print('Accuracy for',sensor,':',correct/(len(data_1)+len(data_2)))
    print('\n')
```

```python
#Door Sensor
compute_door_sensor('door_sensor1','r1','r2')
compute_door_sensor('door_sensor2','r3','c1')
compute_door_sensor('door_sensor3','r4','c1')
compute_door_sensor('door_sensor4','r4','r5')
compute_door_sensor('door_sensor5','r5','r6')
compute_door_sensor('door_sensor6','r9','c2')
compute_door_sensor('door_sensor7','r10','c2')
compute_door_sensor('door_sensor8','r7','r8')
compute_door_sensor('door_sensor9','r7','c3')
compute_door_sensor('door_sensor10','r2','c3')
compute_door_sensor('door_sensor11','r10','c3')
```
