```python
data = pd.read_csv('sorted_data.csv')

graph_structure = {
    'r1': ['r2'], 
    'r2': ['r1', 'c3'], 
    'r3': ['outside', 'c1', 'c2'], 
    'r4': ['r5', 'c1', 'c2'], 
    'r5': ['r6', 'r4'], 
    'r6': ['r5'], 
    'r7': ['c3', 'r8'], 
    'r8': ['r7'], 
    'r9': ['c1', 'c2'], 
    'r10': ['c3', 'c1', 'c2'], 
    'c1': ['r3', 'r4', 'r9'], 
    'c2': ['r3', 'r4', 'r9', 'r10'], 
    'c3': ['r2', 'r7', 'r10'], 
    'outside': ['r3']
}

def get_potential_destinations(sensor_data, graph):
    potential_destinations = {room: set() for room in graph}
    for room, connected_rooms in graph.items():
        motion_sensor = sensor_data.get(f'motion_sensor{room[1:]}')
        door_sensor = sensor_data.get(f'door_sensor{room[1:]}', 0)
        if motion_sensor == 'motion' or door_sensor > 0:
            potential_destinations[room].update(connected_rooms)
        for robot_id in ['robot1', 'robot2']:
            robot_data = sensor_data.get(robot_id)
            if robot_data:
                robot_room, count = eval(robot_data) 
                if robot_room == room and count > 0:
                    potential_destinations[room].update(connected_rooms)
    return potential_destinations

def calculate_potential_destinations(df, graph):
    potential_destinations_time = {}
    for index, row in df.iterrows():
        sensor_data = row.to_dict()
        potential_destinations = get_potential_destinations(sensor_data, graph)
        potential_destinations_time[sensor_data['time']] = potential_destinations
    return potential_destinations_time

potential_destinations_data = calculate_potential_destinations(data, graph_structure)

print('Data First Entry Potential Destinations:')
print(potential_destinations_data[next(iter(potential_destinations_data))])
```

```python
outcomeSpace = {
    'r1': ['r2'], 
    'r2': ['r1', 'c3'], 
    'r3': ['outside', 'c1', 'c2'], 
    'r4': ['r5', 'c1', 'c2'], 
    'r5': ['r6', 'r4'], 
    'r6': ['r5'], 
    'r7': ['c3', 'r8'], 
    'r8': ['r7'], 
    'r9': ['c1', 'c2'], 
    'r10': ['c3', 'c1', 'c2'], 
    'c1': ['r3', 'r4', 'r9'], 
    'c2': ['r3', 'r4', 'r9', 'r10'], 
    'c3': ['r2', 'r7', 'r10'], 
    'outside': ['r3']
}

def count_tran(data, tar):
    d1 = data[tar].to_numpy()
    d2 = np.roll(d1, -1)
    if d1.sum() == 0:
        return 0
    changes = np.maximum(d1 - d2, 0)
    return changes.sum() / d1.sum()

def calculate_transition_matrix(data, period_data, index, outcomeSpace):
    transition_matrix = {}
    
    for key in outcomeSpace.keys():
        total_count = period_data[key].sum()
        if total_count == 0:
            continue  # Avoid division by zero
        
        transitions = np.zeros(len(index))
        
        for i in period_data.index[:-1]:
            t1_d = period_data.loc[i]
            t2_d = period_data.loc[i + 1]
            diff = max(0, t1_d[key] - t2_d[key])
            
            for mov in outcomeSpace[key]:
                diff_m = max(0, t2_d[mov] - t1_d[mov])
                to_add = min(diff, diff_m)
                transitions[index[mov]] += to_add
                diff -= to_add
                if diff <= 0:
                    break
        
        transitions = transitions / total_count
        transitions[index[key]] = 1 - count_tran(period_data, key)
        transition_matrix[key] = transitions

    return transition_matrix

def create_time_periods(data, slices):
    periods = []
    for start, stop in slices:
        periods.append(data.iloc[start:stop])
    return periods

# Prepare the index mapping
index = {key: idx for idx, key in enumerate(outcomeSpace.keys())}

# Read data
data = pd.read_csv('sorted_data.csv')

# Define the time slices for each period
time_slices = [(0, 38), (38, 1918), (1918, 2878), (2878, 4558), (4558, len(data))]
periods = create_time_periods(data, time_slices)

# Calculate transition matrices for each period and store in a dictionary
output = {}
for i, period_data in enumerate(periods, start=1):
    transition_matrix = calculate_transition_matrix(data, period_data, index, outcomeSpace)
    for key, transitions in transition_matrix.items():
        output[f"{key}_s{i}"] = transitions

# Convert the dictionary to a DataFrame and save to CSV
state_csv = pd.DataFrame(output)
state_csv.to_csv('state_matrix_6.csv', index=False)
```
